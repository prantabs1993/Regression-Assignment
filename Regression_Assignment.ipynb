**1.What is Simple Linear Regression?** Ans. Simple Linear Regression is a fundamental statistical technique used to model the relationship between two variables—one independent variable ((X)) and one dependent variable ((Y)). It assumes a linear relationship between these variables, meaning that changes in (X) correspond to proportional changes in (Y). The goal is to find the best-fitting straight line, called the regression line, that minimizes the difference between the actual and predicted values.
Formula:
[ Y = \beta_0 + \beta_1 X + \epsilon ] where:
- (Y) is the dependent variable (target),
- (X) is the independent variable (predictor),
- (\beta_0) is the intercept (value of (Y) when (X = 0)),
- (\beta_1) is the slope (rate of change of (Y) with respect to (X)),
- (\epsilon) is the error term (random noise).


**2.What are the key assumptions of Simple Linear Regression?**
Ans.Simple Linear Regression relies on several fundamental assumptions to ensure accurate predictions and meaningful interpretations. Here are the key ones:
1. Linearity
- The relationship between the independent variable ((X)) and the dependent variable ((Y)) must be linear.
- If the data exhibits a curved trend, a linear model may not be appropriate.
2. Independence
- Observations should be independent of each other.
- If data points are related (e.g., time-series data with dependencies), a different model like Autoregressive models may be more suitable.
3. Homoscedasticity (Constant Variance)
- The variance of residuals (errors) should be constant across all values of (X).
- If residuals increase or decrease systematically, heteroscedasticity exists, and techniques like transformations (log, square root) can help.
4. Normality of Residuals
- The residuals (differences between predicted and actual values) should ideally follow a normal distribution.
- If residuals are skewed, the predictions may not be reliable for inference.
5. No Multicollinearity (Not applicable for Simple Linear Regression, but relevant for Multiple Regression)
- In Multiple Linear Regression, independent variables should not be highly correlated with each other.
- In Simple Linear Regression, this assumption is not a concern since there’s only one predictor.



**3. What does the coefficient m represent in the equation Y=mX+c?**Ans. Slope of a straight line on a 2d plane with x and y representing the horizontal and the vertical axes. It is the change in y with respect to the change in x.



**4. What does the intercept c represent in the equation Y=mX+c?**Ans. Where the straight line in the x,y plane cuts the y axis. In short, it is the value of y when x is 0.



**5.How do we calculate the slope m in Simple Linear Regression?**Ans.With two data points in the x-y plane as (x1,y1)&(x2,y2), slope is (y2-y1)/(x2-x1).



**6.What is the purpose of the least squares method in Simple Linear Regression?**Ans.The least squares method in simple linear regression is used to find the line of best fit that minimizes the sum of the squared differences between the actual data points and the predicted values on the line. This line, also known as the least squares regression line, represents the relationship between the independent and dependent variables in the data.

How is the coefficient of determination (R²) interpreted in Simple Linear Regression?**Ans.The coefficient of determination, denoted as ( R^2 ), is a key metric in Simple Linear Regression that evaluates the goodness of fit—essentially, how well the model explains the variability in the dependent variable.
Interpretation of ( R^2 ):
- Range: ( R^2 ) values range from 0 to 1.
- ( R^2 = 0 ) → The model explains none of the variance in the dependent variable.
- ( R^2 = 1 ) → The model explains all the variance perfectly.
- Proportion of Explained Variance: If ( R^2 = 0.85 ), it means 85% of the variation in the dependent variable is explained by the independent variable, while the remaining 15% is due to other factors or noise.
Limitations:
- Doesn't imply causation: A high ( R^2 ) doesn’t guarantee a causal relationship.
- Doesn't measure model complexity: A high ( R^2 ) could indicate overfitting, especially when additional variables are included in multiple regression.
- Sensitive to outliers: Extreme values can distort ( R^2 ), making the model appear better or worse than it actually is.



**8. What is Multiple Linear Regression?**Ans.Multiple Linear Regression (MLR) is an extension of Simple Linear Regression, where instead of using just one independent variable to predict the dependent variable, multiple independent variables are used.
Formula for MLR:
[ Y = \beta_0 + \beta_1X_1 + \beta_2X_2 + ... + \beta_nX_n + \epsilon ] where:
- ( Y ) is the dependent variable (target).
- ( X_1, X_2, ..., X_n ) are independent variables (predictors).
- ( \beta_0 ) is the intercept.
- ( \beta_1, \beta_2, ..., \beta_n ) are the regression coefficients.
- ( \epsilon ) is the error term.


**9.What is the main difference between Simple and Multiple Linear Regression?**Ans.The key difference between Simple Linear Regression and Multiple Linear Regression lies in the number of independent variables used to predict the dependent variable. In Simple Linear Regression the number of independent variables is one, while it may be multiple in the case of Multiple Linear Regression.


**10.What are the key assumptions of Multiple Linear Regression?**Ans.Key Assumptions of MLR:
- Linearity
- The relationship between the independent variables ((X_1, X_2, ..., X_n)) and the dependent variable ((Y)) should be linear.
- This can be checked using scatter plots or residual plots.
- Independence
- Observations should be independent of each other.
- If the data points are correlated (as in time-series data), techniques like autoregressive models should be considered.
- Homoscedasticity (Constant Variance of Errors)
- The variance of residuals should remain constant across all levels of the independent variables.
- Heteroscedasticity can be detected using residual vs. fitted value plots and corrected with transformations like log or Box-Cox.
- No Multicollinearity
- Independent variables should not be highly correlated with each other.
- High multicollinearity distorts coefficient estimates, making them unreliable.
- This can be assessed using the Variance Inflation Factor (VIF) and addressed by removing or combining correlated predictors.
- Normally Distributed Errors
- Residuals should be normally distributed to ensure reliable hypothesis testing and confidence intervals.
- Can be checked using a Q-Q plot or Shapiro-Wilk test. If violated, transformations like log-scaling can help.


**11.What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?**Ans.Heteroscedasticity occurs when the variance of residuals (errors) is not constant across all levels of the independent variables. Instead of having a uniform spread, residuals may exhibit increasing or decreasing variance, which can distort statistical inference.
Effects on Multiple Linear Regression:
- Biased Standard Errors
- Since standard errors are inflated or underestimated, confidence intervals become unreliable.
- This affects hypothesis tests, making it harder to determine the significance of predictor variables.
- Inefficient Estimates
- Ordinary Least Squares (OLS) assumes homoscedasticity. When violated, OLS still provides unbiased coefficient estimates but these estimates are no longer efficient (i.e., they have larger variances).
- Incorrect p-values
- T-tests and F-tests rely on assumptions of homoscedasticity.
- If heteroscedasticity is present, p-values may be misleading, leading to incorrect conclusions about variable significance.
- Distorted Model Interpretation
- If variance grows with higher values of independent variables, predictions become less reliable for extreme values.
- Outliers or extreme points can disproportionately affect the regression results.


**12.How can you improve a Multiple Linear Regression model with high multicollinearity?**Ans.High multicollinearity in a Multiple Linear Regression (MLR) model can distort coefficient estimates, making it difficult to interpret variable importance and leading to unreliable predictions. Here’s how you can improve the model:
1. Check for Multicollinearity
- Use Variance Inflation Factor (VIF) to quantify the severity.
- If VIF > 5 or 10, multicollinearity is considered problematic.
2. Feature Selection
If two variables are highly correlated, consider:
- Dropping one of them if both contain similar information.
- Using Principal Component Analysis (PCA) to reduce dimensionality.
- Applying Lasso Regression to shrink less important variables.
3. Transform Variables
Sometimes, applying log transformations or differencing can reduce collinearity.
4. Use Ridge or Lasso Regression
- Ridge Regression (L2 regularization) reduces the impact of collinear predictors.
- Lasso Regression (L1 regularization) can shrink coefficients toward zero.
from sklearn.linear_model import Ridge
5. Use Partial Least Squares (PLS)
PLS regression finds latent components that explain variance while avoiding collinearity issues.

**15.How can the interpretation of intercept differ between Simple and Multiple Linear Regression**Ans.Simple Linear Regression:
The intercept (b0) in simple linear regression, y = b0 + b1x, represents the predicted value of the dependent variable (y) when the independent variable (x) is equal to zero.
For example, if you're predicting salary based on years of experience (x), the intercept would be the predicted salary for someone with zero years of experience.
Multiple Linear Regression:
The intercept (b0) in multiple linear regression, y = b0 + b1x1 + b2x2 + ... + bnxn, represents the predicted value of the dependent variable (y) when all independent variables (x1, x2, ..., xn) are equal to zero.
This means that the intercept represents the average outcome when all predictor variables are at their zero value.
For instance, if you're predicting house price based on square footage, number of bedrooms, and lot size, the intercept would be the predicted house price when all those variables are zero.

**20.How can heteroscedasticity be identified in residual plots, and why is it important to address it.**Ans.Heteroscedasticity, or non-constant variance of errors, in residual plots is typically identified by a "fan-shaped" or "cone-shaped" pattern, where the spread of residuals increases or decreases systematically with the fitted values. Addressing heteroscedasticity is important because it violates a key assumption of Ordinary Least Squares (OLS) regression and can lead to inaccurate standard errors and unreliable statistical inferences.

**23.What is polynomial regression?**Ans.Polynomial regression is an extension of linear regression used to model non-linear relationships between variables. **24.How does polynomial regression differ from linear regression?**Ans.Polynomial regression extends linear regression by allowing for non-linear relationships between variables. **25.When is polynomial regression used?**Ans.Polynomial regression is used when the relationship between an independent variable and a dependent variable is not linear, but rather curved or non-linear.


**26.What is the general equation for polynomial regression?**Ans.The general equation for polynomial regression, a method used to fit a polynomial of a specific degree to a dataset, is:
y = β₀ + β₁x + β₂x² + ... + βₙxⁿ + ε
where:
y: is the dependent variable.
x: is the independent variable.
β₀, β₁, β₂, ... βₙ: are the coefficients of the polynomial terms (the parameters to be estimated).
n: is the degree of the polynomial.
ε: is the error term.


**27.Can polynomial regression be applied to multiple variables**Ans.Yes, polynomial regression can be applied to multiple variables.
```python
#31.How is polynomial regression implemented in Python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline

       # Example Data
x = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 6, 12, 20, 30])
degree = 2 # Example degree of the polynomial
poly_features = PolynomialFeatures(degree=degree)
x_poly = poly_features.fit_transform(x)


model = LinearRegression()
model.fit(x_poly, y)

x_plot = np.linspace(x.min(), x.max(), 100).reshape(-1, 1)
x_plot_poly = poly_features.transform(x_plot)
y_plot = model.predict(x_plot_poly)
plt.scatter(x, y, color='blue', label='Data Points')
plt.plot(x_plot, y_plot, color='red', label='Polynomial Regression Curve')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
